```python
class Animals():
  def __init__(self, name, weight, voice):
    self.name = name
    self.weight = weight
    self.voice = voice

  def h(self):
    print("Hi, this is Joe's farm!")

class Geese(Animals):
  def feeding(self):
    self.n = float(input("Enter normal weight: "))
    print("Name:", self.name)
    if self.weight > self.n:
      print("Overweight! You need to put feed 100 grams less.")
    elif self.weight < self.n:
      print("Underweight! Put 100 grams more feed!")

  def vc(self):
    print("Geese voice:", self.voice)

  def weight_count(self):
    self.res.append(self.weight)

class Cows(Animals):
  def feeding(self):
    self.n = float(input("Enter normal weight: "))
    print("Name:", self.name)
    if self.weight > self.n:
      print("Overweight! You need to put feed 100 grams less.")
    elif self.weight < self.n:
      print("Underweight! Put 100 grams more feed!")

  def milk(self):
    try:
      self.t = int(input("Enter the number of cows milked today: "))
    except:
      print("Enter an integer!")
    print("Name:", self.name)
    print("Number of cow milked today:", self.t)

  def vc(self):
    print("Cows voice:", self.voice)

  def weight_count(self):
    self.res.append(self.weight)

class Sheep(Animals):
  def feeding(self):
    self.n = float(input("Enter normal weight: "))
    if self.weight > self.n:
      print("Overweight! You need to put feed 100 grams less.")
    elif self.weight < self.n:
      print("Underweight! Put 100 grams more feed!")

  def haircut(self):
    self.a = float(input("Enter the amount of wool obtained in grams: "))
    print("Name:", self.name)
    print("You got wool:", self.a)

  def vc(self):
    print("Sheep voice:", self.voice)

  def weight_count(self):
    self.res.append(self.weight)

class Chicken(Animals):
  def feeding(self):
    self.n = float(input("Enter normal weight: "))
    print("Name:", self.name)
    if self.weight > self.n:
      print("Overweight! You need to put feed 100 grams less.")
    elif self.weight < self.n:
      print("Underweight! Put 100 grams more feed!")

  def egg_collection(self):
    try:
      self.e = int(input("Enter the rate of eggs per month: "))
      self.e2 = int(input("Enter the number of eggs you collected per month: "))
    except:
      print("Enter an integer!")
    if self.e > self.e2:
      print("You didn't meet the standard! Collect more eggs.")
    elif self.e < self.e2:
      print("You completed the plan!")
  
  def vc(self):
    print("Chiken voice:", self.voice)

  def weight_count(self):
    self.res.append(self.weight)

class Goats(Animals):
  def feeding(self):
    self.n = float(input("Enter normal weight: "))
    print("Name:", self.name)
    if self.weight > self.n:
      print("Overweight! You need to put feed 100 grams less.")
    elif self.weight < self.n:
      print("Underweight! Put 100 grams more feed!")
  
  def vc(self):
    print("Goats voice:", self.voice)

  def weight_count(self):
    self.res.append(self.weight)

class Ducks(Animals):
  def feeding(self):
    self.n = float(input("Enter normal weight: "))
    print("Name:", self.name)
    if self.weight > self.n:
      print("Overweight! You need to put feed 100 grams less.")
    elif self.weight < self.n:
      print("Underweight! Put 100 grams more feed!")

  def vc(self):
    print("Ducks voice:", self.voice)

  def weight_count(self):
    self.res.append(self.weight)
```

```python
goose1 = Geese("Gray", 3, "Ga")
goose1.feeding()
goose1.vc()

cow1 = Cows("Manka", 12, "Mu")
cow1.feeding()
cow1.milk()
cow1.vc()

sheep1 = Sheep("Barashek", 15, "Be")
sheep1.feeding()
sheep1.haircut()
sheep1.vc

chicken1 = Chicken("Co-co", 3, "Co")
chicken1.feeding()
chicken1.egg_collection()
chicken1.vc()

goat1 = Goats("Roga", 5, "Me")
goat1.feeding()
goat1.vc()

duck1 = Ducks("Krakva", 2, "Krya")
duck1.feeding()
duck1.vc()

total_weight = goose1.weight + cow1.weight + sheep1.weight + chicken1.weight + goat1.weight + duck1.weight 
print(total_weight)

animals = [goose1.weight, cow1.weight, sheep1.weight, chicken1.weight, goat1.weight, duck1.weight]
print(max(animals))
```
