```python
import math

def f(x, option):
    if option == 1:
        return math.sin(x)
    elif option == 2:
        return x**2
    elif option == 3:
        return math.exp(x)
    else:
        raise ValueError("Некорректный выбор функции.")

def w(x, y, option):
    fx = f(x, option)
    if x >= y:
        return (x - y)**3 + math.atan(fx)
    else:
        fy = f(y, option)
        return fx / (y + fy**2) - 0.5

def main():
    print("Выберите функцию f(x):")
    print("1. sin(x)")
    print("2. x^2")
    print("3. exp(x)")

    option = int(input("Введите номер функции (1, 2 или 3): "))

    x = float(input("Введите значение x: "))
    y = float(input("Введите значение y: "))

    try:
        result = w(x, y, option)
        print(f"Результат w({x}, {y}) = {result}")
    except ValueError as e:
        print(e)

if __name__ == "__main__":
    main()
```

```python
import math
from prettytable import PrettyTable

def series_s(x, n):
    """
    Вычисление суммы степенного ряда S(x) с n членами.
    """
    s = 0
    for k in range(n + 1):
        term = ((-1)**k * x**(2 * k)) / math.factorial(2 * k)
        s += term
    return s

def calculate_error(x, n):
    """
    Вычисление погрешности между рядом S(x) и функцией cos(x).
    """
    actual_value = math.cos(x)
    approx_value = series_s(x, n)
    error = abs((actual_value - approx_value) / actual_value) * 100
    return error

def main():
    table = PrettyTable()
    table.field_names = ["x", "S(x) при n=3", "S(x) при n=5", "cos(x)", "Ошибка при n=3 (%)", "Ошибка при n=5 (%)"]

    x = 0.0
    while x <= 1.0:
        s_n3 = series_s(x, 3)
        s_n5 = series_s(x, 5)
        actual_cos = math.cos(x)
        error_n3 = calculate_error(x, 3)
        error_n5 = calculate_error(x, 5)

        table.add_row([f"{x:.1f}", f"{s_n3:.6f}", f"{s_n5:.6f}", f"{actual_cos:.6f}", f"{error_n3:.2f}", f"{error_n5:.2f}"])
        x += 0.1

    print(table)

if __name__ == "__main__":
    main()
```

```python
import math
import cmath

def calculate_expressions():
    # Первое выражение
    x1 = 0.54
    expr1 = (abs(x1 - 1)**0.25) + math.log(x1**3)
    print(f"1. Результат первого выражения: {expr1:.6f}")

    # Второе выражение
    x2, y2 = -10.5, 20.5
    expr2 = ((x2 + y2) * math.sqrt(y2 + 10.5)) / (x2**2 - y2)
    print(f"2. Результат второго выражения: {expr2:.6f}")

    # Третье выражение
    x3 = math.radians(78.5)  # Перевод угла в радианы
    y3 = math.radians(25)    # Перевод угла в радианы
    z3 = 3.2
    expr3 = (math.sin(x3**3) + math.sqrt(x3 + y3 + 50) + math.atan(z3)) / math.cos(y3**2)
    print(f"3. Результат третьего выражения: {expr3:.6f}")

    # Четвёртое выражение
    x4 = complex(-5.7, 1.22)
    y4 = complex(10, 3.7)
    expr4 = cmath.exp(x4) - cmath.sqrt(x4) + y4
    print(f"4. Результат четвёртого выражения: {expr4:.6f}")

def main():
    calculate_expressions()

if __name__ == "__main__":
    main()
```

```python
import math

# Запрос значений x, y, z у пользователя
x = float(input("Введите значение x: "))
y = float(input("Введите значение y: "))
z = float(input("Введите значение z: "))

# Вычисление выражения
try:
    numerator = 1 + z + (x ** 2) / (2 * y) + (z * math.sqrt(abs(y))) / (3 * x)
    expression = abs(math.pi + 3 * x * (math.cos(5 * x))) * numerator
    print(f"Результат выражения: {expression}")
except ZeroDivisionError:
    print("Ошибка: Деление на ноль. Проверьте введённые значения x и y.")
except ValueError as e:
    print(f"Ошибка вычислений: {e}")
```
