```python
def is_leap(year):
  if year % 4 == 0 and year % 100 != 0 or year % 400 == 0:
    febrary_day=29
  else:
     febrary_day = 28
  return febrary_day

def days(month, year):
  month_day_31=[1,3,5,7,8,10,12]
  month_day_30=[4,6,9,11]
  if month in month_day_31:
    date=31
  elif month in month_day_30:
    date=30
  else:
    date=is_leap(year)
  return date

def previous_date(day, month, year):
  if day == 1:
    month -=1
    if month == 0:
      year -=1
      month=12
      day=days(month, year)
    else:
      year=year
      month = month
      day=days(month, year)
  else:
    day -=1
    month=month
    year=year
  return day, month, year

def next_date(day, month, year):
  if day == days(month, year):
    month += 1
    if month == 13:
      month=1
      year+=1
      day=1
    else:
      month=month
      day=1
      year=year
  else:
    day+=1
    month = month
    year=year
  return day, month, year

day, month, year = input('День, месяц, год через пробел: ').split()
day=int(day)
month=int(month)
year=int(year)
print("Предыдущий день: ", previous_date(day, month, year))
print("Следующий день: ", next_date(day, month, year))
# --------------
# Пример вывода:
#
# День, месяц, год через пробел: 1 3 2000
# Предыдущий день: 29/02/2000
# Следующий день: 02/03/2000
```

# № 5.2.14

Напишите функцию, которая принимает неограниченное количество числовых аргументов и возвращает кортеж из двух списков:

отрицательных значений (отсортирован по убыванию);

неотрицательных значений (отсортирован по возрастанию).

```python
def split_numbers(*args):
  pos=[]
  neg=[]
  for i in args:
    if i > -1:
      pos.append(i)
    elif i < 0:
      neg.append(i)
    pos.sort()
    neg.sort(reverse = True)
  return pos,neg

print(split_numbers(1, 4, -5, 0, -33))

# --------------
# Пример вывода:
#
# ([-5, -33], [0, 1, 4])

```

суммы цифр числа;

количества цифр в числе.

```python
def digits_sum(value):
  if value == 0:
    return 0
  else:
    return (value%10) + digits_sum(value//10) 



def digits_count(value):
  if value == 1:
    return 1
  else:
    return value**0 + digits_count(value//10)

print(digits_sum(12345))
print(digits_count(12345))

# --------------
# Пример вывода:
#
# 15
# 5
```
